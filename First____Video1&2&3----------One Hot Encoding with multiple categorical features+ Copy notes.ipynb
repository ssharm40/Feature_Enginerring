Suppose you have a dataset in which where you have one feature with 100 categorical values so when you perform one hot encoding then it will create 100 columns

So for handling this things we use below technique

```python
df= pd.read_csv('Data.csv', usecols=['X1','X2','X3','X4','X5','X6'])
df.head()
```

```python
# Now we want to find how many categories each columnn has



for col in df.columns:
    print(col, ': ', len(df[col].unique()), 'Lables')
    
'''    

len(df.X2.unique())
'''
```

```python
# So there are many categories if we use one hot encoding then this will be the size of data: 

#One hot encoding

temp= pd.get_dummies(df, drop_first=True)
#temp.head()
temp.shape

#Shape is (4209, 117) means 117 coulmns- so these are very big amount of columns here we need to 
# perform some special encoding technique
```

We can see that from just 6 initial categorical features, we end up with 117 new variables

Now what should we do? 

In 2017,there was a kaggle challange to solve this kind of problem and winner give the below solution

They kept top 10 repeated features for other they just keept it as zero or consider it as outliers

How to do this with python,let understand this below : 




```python
# Let's find the top 10 most frequent feature categories for columns value X2

df.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
# Let's make a list with the most frequent categories of the variable: 

top_10= [x for x in df.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
# Now we make the 10 binary variables


for label in top_10:
    df[label]= np.where(df['X2']==label , 1, 0)
df

# Now it's done for X2 table
```

```python
# Now we apply these for all


def one_hot_top_x(df, variable, top_x_lables):
    # Function to create dummy variables for the most frequent lables, 
    #We can vary the no of most frequent lables that we encode.
    
    for label in top_x_lables:
        df[variable+'_'+label]= np.where(df[variable]==label,1,0)

        
#Reading the data again : 
df= pd.read_csv('Data.csv', usecols=['X1','X2','X3','X4','X5','X6'])

#Encode x2 into the top 10 most frequent categories:

one_hot_top_x(df, 'X2', top_10)
#df.head()
```

```python
# Now we can done for all at the same time with only 5 most category
df= pd.read_csv('Data.csv', usecols=['X1','X2','X3','X4','X5','X6'])

top_5=[]
for i in df.columns:
    temp= [x for x in df[i].value_counts().sort_values(ascending=False).head(5).index]
    top_5.append(temp)
j=0
for i in df.columns:
    print(i)
    one_hot_top_x(df,i , top_5[j])
    j=j+1

    
df
    
```

*Two type of Categorical Data*

1) Nominal Encoding--- Where we don't care about Order
                       Exp- Gender Columns(Male or Female), State Column(UP, UK, MP..)
                       So we don't need to care about the order here


2) Ordinal Encoding--- We need to care about Rank
                       Exp- like education(BE, BCOM, PHD, Master)
                       so you need to rank them like this------ PHD-1, Master-2, BE-3, Bcom-4
