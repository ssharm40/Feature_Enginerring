#### So as it's create many more feature so one hot encoding technique is fail here so we have to apply *One hot encoding with multiplefeatures*

```python
df=pd.read_csv('Data.csv', usecols=['X0','X1','X2','X3','X4','X5','X6'])
```

```python
df.head()
```

```python
pd.get_dummies(df, drop_first=True).shape

# So that'shuge number
```

```python
#Taking max frequency

df.X1.value_counts().sort_values(ascending=False).head(10)
```

```python
# Now we require the index name

list_10= list(df.X1.value_counts().sort_values(ascending=False).head(10).index)
list_10
```

```python
# Now apply one hot encoding with these top 10 features

for categories in list_10:
    df[categories]= np.where(df['X1']==categories, 1,0)
```

```python
list_10.append('X1')
df[list_10].head(10)
```

## ---------------------- 3rd Tech- Ordinal number encoding

Where you rank the order.

**Exp**--- Grading system, 

Explaination
Suppose we have a column with name of weekdays, and we are trying to find out which employee is more happy.
['Monday', 'Tuesday'..........,'Sunday'] so we want to encode it in this format{'Monday': 1, 'Tuesday':2,...........}

here we will create this dictonary for ordering 
now we use the pandas map() function to map these ordering


```python
# Taking the example of datetime
import datetime
```

```python
today_date= datetime.datetime.now()
today_date
```

```python
# Now we create 20s day data

today_date-datetime.timedelta(2)
```

```python
##List Comprehension

top_20= [today_date-datetime.timedelta(x) for x in range(0,20,1)]
```

```python
df=pd.DataFrame(top_20)
df
```

```python
df.columns=['Day']
df.head()
```

```python
df['weekday']= df['Day'].dt.day_name()
```

```python
df.head()
```

```python
# For encoding we create dictionary

dictionary={'Monday':1, 'Tuesday':2, 'Wednesday':3, 'Thursday':4, 'Friday':5, 'Saturday':6, 'Sunday':7}
dictionary
```

```python
# Now we apply map fucntion to encode our day column

df['Weekday_enocding']= df['weekday'].map(dictionary)
df.head()

```

## Next Tech----------------------- Count or Frequency Encoding


In This method category will be replaced by no of count.

Exp: In below example we have column with name country, which contain multiple category.
so we find the no of occurance for each category by .value_count() method
and then map these sum of count with country name

```python
df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', usecols=[1,3,5,6,7,8,9,13], header = None,index_col=None)
```

```python
df.head()
```

```python
len(df[1].unique())
```

```python
# Let'stake categorical feature in one place

col=[1,3,5,6,7,8,9,13]
```

```python
df[col].head()
```

```python
# Let's give them name

df.columns=['Employement', 'Degree', 'Status', 'Designation', 'Family_job', 'Race', 'Sex', 'Country']
df.head()
```

```python
df.columns.to_list()
```

```python
# To Find out unique category for each feature

for cols in df.columns.to_list():
    print(cols, ': ',len(df[cols].unique()))
```

```python
df.Country.value_counts().head(12)

#For question mark we can encode by ranom sampel any other tech 
```

```python
# Now creation dictionary

d1= df.Country.value_counts().to_dict()
d1
#As this tech replace it by no of count
#Exp- United state will be replaced by 29170, Cuba-95
```

```python
df['Country_map']= df['Country'].map(d1)
```

```python
df.head(20)
```

**Advantage**
1) Easy to Use

2) Not increasing any feature space as we will delete the previous column



**Disadvantage**
1) What if two country have same coun so it will not be able to handle this. It provide the same weight
